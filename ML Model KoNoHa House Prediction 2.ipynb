###Importing Essential Libraries for Data Analysis and Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

###Load Data from a CSV File into a Pandas DataFrame

```python
data = pd.read_csv('/content/220824_DataClean_B.csv')
```

###Displaying the First Few Rows of the Dataset

```python
data.head()
```

###Displaying Information About the Dataset

```python
data.info()
```

###Counting Missing Values in Each Column of the Dataset

```python
data.isnull().sum()
```

###Filling Missing Values in 'Store' Column with Zero

```python
#Handle missing data 'Store'
data['Store'].fillna(0, inplace=True)
```

###Counting Missing Values in Each Column of the Dataset

```python
data.isnull().sum()
```

###Generating Descriptive Statistics for the Dataset

```python
data.describe()
```

###Removing Infinite Values from the 'Price/Sqft' Column

```python
#Handle infinite values in 'Price/Sqft'
data = data[data['Price/Sqft'] != np.inf]
```

###Removing Rows with Missing Values from the Dataset

```python
#Drop any missing values in any rows
data.dropna(inplace=True)
```

###Generating Descriptive Statistics for the Dataset

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

###Selecting Features and Target Variable for Modeling

```python
# Select features and target variable
X = data[['Location', 'Rooms', 'Store', 'Bathrooms', 'Property Type', 'Size (sqft)', 'Furnishing']]
y = data['Price (RM)']
```

###Splitting Data into Training and Testing Sets

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

###Training a Linear Regression Model

```python
# Train a Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

###Making Predictions with a Trained Model on the Test Set

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

###Model Performance Evaluation Using MSE and RÂ² Metrics

```python
# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

#Output
print(f"MSE: {mse}")
print(f"RÂ²: {r2}")
```

###Training a Random Forest Regressor Model

```python
from sklearn.ensemble import RandomForestRegressor

# Train a Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

###Random Forest Model Prediction and Performance Evaluation

```python
# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)

# Evaluate the model's performance
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

mse_rf, r2_rf
```

###Training an XGBoost Regressor Model

```python
import xgboost as xgb

# Train an XGBoost model
xgboost_model = xgb.XGBRegressor(n_estimators=100, random_state=42)
xgboost_model.fit(X_train, y_train)
```

###XGBoost Model Prediction and Performance Evaluation

```python
# Make predictions on the test set
y_pred_xgb = xgboost_model.predict(X_test)

# Evaluate the model's performance
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)

mse_xgb, r2_xgb
```

###Training a Decision Tree Regressor Model

```python
from sklearn.tree import DecisionTreeRegressor

# Initialize and train a Decision Tree model
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
```

###Decision Tree Model Prediction and Performance Evaluation

```python
# Make predictions on the test set
y_pred_dt = dt_model.predict(X_test)

# Evaluate the model's performance
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

mse_dt, r2_dt
```

###Model Performance Evaluation and Comparison Using Common Regression Metrics

```python
#Comparing model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.linear_model import LinearRegression

# Assume y_test is the true values and y_pred_* are predictions from different models
def evaluate_model(y_test, y_pred, model_name):
    mse = mean_squared_error(y_test, y_pred)
    rmse = mse ** 0.5
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"{model_name} Performance:")
    print(f"  MSE: {mse}")
    print(f"  RMSE: {rmse}")
    print(f"  MAE: {mae}")
    print(f"  RÂ²: {r2}\n")

# Train a Linear Regression model and make predictions (if you haven't already)
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)

# Example: Comparing models
evaluate_model(y_test, y_pred_lr, "Linear Regression")
evaluate_model(y_test, y_pred_rf, "Random Forest")
evaluate_model(y_test, y_pred_xgb, "XGBoost")
evaluate_model(y_test, y_pred_dt, "Decision Tree")
```

###Evaluating Model Generalization by Comparing Training and Testing Performance

```python
#Compare between training and testing performance
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

# Set a threshold for acceptable difference (example: 0.1)
threshold = 0.1

# Add the model name here
model_name = "Linear Regression"  # Replace with the actual model name

if test_score > train_score - threshold:
    print(f"Model {model_name} generalizes well")
```

###Selecting the Best Model Based on RÂ² Score

```python
#Determine the best model
models = {
    'Linear Regression': lr_model,
    'Random Forest': rf_model,
    'XGBoost': xgboost_model,
    'Decision Tree': dt_model
}

best_model = None
best_r2 = -float('inf')

for model_name, model in models.items():
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    if r2 > best_r2:
        best_r2 = r2
        best_model = model_name

print(f"The best model is {best_model} with RÂ²: {best_r2}")
```

###Comparison of Model Predictions vs. Actual Values



```python
#Comparison of model predictions
plt.figure(figsize=(10, 6))

# Plot Linear Regression predictions
plt.scatter(y_test, y_pred_lr, alpha=0.5, color='blue', label='Linear Regression')

# Plot Random Forest predictions
plt.scatter(y_test, y_pred_rf, alpha=0.5, color='green', label='Random Forest')

# Plot XGBoost predictions
plt.scatter(y_test, y_pred_xgb, alpha=0.5, color='purple', label='XGBoost')

# Plot the perfect prediction line
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Perfect Prediction')

# Add labels and title
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Comparison of Model Predictions')
plt.legend()
plt.show()
```
